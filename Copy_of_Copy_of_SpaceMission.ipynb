```python


# Sample space missions dataset
data = {
    'Location': ['Guiana Space Centre, French Guiana, France','ELA-3, Guiana Space Centre, French Guiana, France','ELA-3, Guiana Space Centre, French Guiana, France','ELA-3, Guiana Space Centre, French Guiana, France','ELA-3, Guiana Space Centre, French Guiana, France','ELA-3, Guiana Space Centre, French Guiana, France','ELA-3, Guiana Space Centre, French Guiana, France','ELA-3, Guiana Space Centre, French Guiana, France','ELA-3, Guiana Space Centre, French Guiana, France','ELA-3, Guiana Space Centre, French Guiana, France'],
    'Year': [1997, 1999, 2003,2004,2005,2006,2007,2008,2009,2010],
}
space_missions = pd.DataFrame(data)

# Group by 'Location' and 'Year' and count the number of missions
grouped_data = space_missions.groupby(['Location', 'Year']).size().unstack(fill_value=0)

# Create a heatmap
plt.figure(figsize=(10, 6))
plt.imshow(grouped_data, cmap='viridis', interpolation='nearest')
plt.colorbar(label='Number of Missions')
plt.xticks(np.arange(len(grouped_data.columns)), grouped_data.columns, rotation=45)
plt.yticks(np.arange(len(grouped_data.index)), grouped_data.index)
plt.xlabel('Year')
plt.ylabel('Location')
plt.title('Number of Space Missions by Location and Year')
```

**3D Scatter Plot**

```python

from mpl_toolkits.mplot3d import Axes3D

# Sample space missions dataset
data = {
    'Location': ['Guiana Space Centre, French Guiana, France','ELA-3, Guiana Space Centre, French Guiana, France','ELA-3, Guiana Space Centre, French Guiana, France','ELA-3, Guiana Space Centre, French Guiana, France','ELA-3, Guiana Space Centre, French Guiana, France','ELA-3, Guiana Space Centre, French Guiana, France','ELA-3, Guiana Space Centre, French Guiana, France','ELA-3, Guiana Space Centre, French Guiana, France','ELA-3, Guiana Space Centre, French Guiana, France','ELA-3, Guiana Space Centre, French Guiana, France'],
    'Year': [1997, 1999, 2003,2004,2005,2006,2007,2008,2009,2010],
    'Price': [20,20,20,200,30,37,25,25,2,2]
}
space_missions = pd.DataFrame(data)

# Generate random payload masses for demonstration
np.random.seed(0)
space_missions['PayloadMass'] = np.random.randint(1000, 10000, size=len(space_missions))

# Create a 3D scatter plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(space_missions['Year'], space_missions['Price'], space_missions['PayloadMass'], c='b', marker='o')

ax.set_xlabel('Year')
ax.set_ylabel('Cost')
ax.set_zlabel('PayloadMass')
ax.set_title('Space Missions: Year vs Cost vs PayloadMass')
plt.show()
```
